```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas
from sklearn.linear_model import LinearRegression
from numpy import log 

df = pandas.read_csv( 'data/assay_data_clean.csv' )
print df.rate.dtype
print df.columns
temp_key = pandas.read_csv( 'data/temp_key.csv' )
temp_key.index = temp_key.Well
temp_key_dict = temp_key.Kelvin.to_dict()
```

```python
R = 8.

# convenience functions 
def fit_mutant( mutant_df ):
    # mutant name 
    name = mutant_df.mutant.unique()[0] # guaranteed to be unique 
    
    # scale rate
    mutant_df['scaled_rate'] = mutant_df.rate / mutant_df.rate.max()
    mutant_df['keq'] = 1 / mutant_df.scaled_rate #ratio of unfolded to folded
    mutant_df['temp'] = mutant_df.well.str[0].map( temp_key_dict ) 
    mutant_df['ln_K'] = log( mutant_df.keq )
    mutant_df['one_over_T'] = 1 / mutant_df.temp
    mutant_df['delta_G'] = R * mutant_df.temp * mutant_df.ln_K

    #Van't Hoff plot 
    plt.scatter( mutant_df.one_over_T, mutant_df.ln_K, alpha=0.7, color='r' )
    plt.title( name )
    plt.xlim( [0.003, 0.0034] )
    plt.xlabel( '$1/T (1/K)$' )
    plt.ylabel( '$ln K$')
    plt.show()
    
    # another plot
    plt.scatter( mutant_df.delta_G, mutant_df.temp, alpha=0.7, color='r' )
    plt.title( name )
    plt.xlabel( '$T (K)$' )
    plt.ylabel( '$\delta G$')
    plt.show()

    # linear regression 
    regr = LinearRegression()

    # fit to logistic model 
    logistic = None
    
    # collect params
    params = {}
    
    return params

def fit_a_mutant_return_params( mutant_df ):
    
    params = fit_mutant( mutant_df )
    print params
    
    return mutant_df
```
