```python
import pandas 
import seaborn 
%matplotlib inline

df = pandas.read_csv( 'assay_data_clean.csv' )
df = df.dropna()
df = df[( df.rate > 0 )]

def percent_rates( dataframe ):
    dataframe.rate = dataframe.rate / dataframe.rate.max() 
    return dataframe 

df.groupby( 'mutant' ).apply( percent_rates )

seaborn.lmplot( 'temp', 'rate', data=df, col='mutant', col_wrap=6, logistic=True )
```
